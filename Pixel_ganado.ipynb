```python
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen
image_path = "/content/dji_fly_20240324_141712_74_1711307854400_photo_optimized.jpg"
image = Image.open(image_path)
draw = ImageDraw.Draw(image)
```

```python
# Obtener dimensiones de la imagen
img_width, img_height = image.size

# Tamaño del grid (10 píxeles por cuadrante)
grid_size = 10
```

```python
# Dibujar líneas del grid
for x in range(0, img_width, grid_size):
    draw.line([(x, 0), (x, img_height)], fill="blue", width=1)
    if x % 100 == 0:  # Agregar texto cada 100 píxeles
        draw.text((x + 5, 5), str(x), fill="red")

for y in range(0, img_height, grid_size):
    draw.line([(0, y), (img_width, y)], fill="blue", width=1)
    if y % 100 == 0:  # Agregar texto cada 100 píxeles
        draw.text((5, y + 5), str(y), fill="red")
```

```python
# Mostrar la imagen con el grid
plt.figure(figsize=(12, 8))
plt.imshow(image)
plt.title("Imagen dividida en cuadrantes con valores de referencia")
plt.axis("off")
plt.show()
```

```python
# --- Seleccionar una región de interés (ROI) y medir ---
# Coordenadas de la región seleccionada (puedes cambiar estas manualmente)
roi_x1, roi_y1, roi_x2, roi_y2 = 750, 450, 850, 550  # Ejemplo de coordenadas

# Dibujar un rectángulo para visualizar la región seleccionada
draw.rectangle([roi_x1, roi_y1, roi_x2, roi_y2], outline="red", width=2)

# Calcular el tamaño en píxeles de la región seleccionada
width_in_pixels = roi_x2 - roi_x1
height_in_pixels = roi_y2 - roi_y1

# Mostrar la imagen con la región seleccionada
plt.figure(figsize=(12, 8))
plt.imshow(image)
plt.title(f"Región seleccionada - Ancho: {width_in_pixels} px, Alto: {height_in_pixels} px")
plt.axis("off")
plt.show()

# Imprimir el tamaño en píxeles de la región
print(f"Ancho de la región seleccionada: {width_in_pixels} píxeles")
print(f"Alto de la región seleccionada: {height_in_pixels} píxeles")
```
